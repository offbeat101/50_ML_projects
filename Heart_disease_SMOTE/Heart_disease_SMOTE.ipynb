```python
from scipy.stats import normaltest
def check_normal_distribution(column):
    stat, p = normaltest(df[column])
    return p > 0.05  # If p > 0.05, we fail to reject the null hypothesis, i.e., data is normally distributed

normality_results = {col: check_normal_distribution(col) for col in df.columns if df[col].dtype != 'object'}
print("Normality Results:\n", normality_results)
```

```python
from scipy.stats import ttest_ind, chi2_contingency


group1 = df1[df1['target'] == 0]['age']
group2 = df1[df1['target'] == 1]['age']
t_statistic, p_value = ttest_ind(group1, group2)
print("T-test for age:")
print("T-statistic:", t_statistic)
print("P-value:", p_value)

contingency_table = pd.crosstab(df['sex'], df['target'])
chi2, p, dof, expected = chi2_contingency(contingency_table)
print("\nChi-square test for sex and target:")
print("Chi-square statistic:", chi2)
print("P-value:", p)
```

T-test for Age:


T-statistic: 4.51
It represents the difference between the means of the two groups (patients with and without heart disease) relative to the variability in the data. A higher t-statistic indicates a larger difference between the groups(suggesting a significant difference in the mean age between patients with and without heart disease).

P-value: 9.04e-06
p-value is very small compared to 0.05. This means we reject the null hypothesis and conclude that there is a statistically significant difference in the mean age between patients with and without heart disease.


---


Chi-square Test for Sex and Target:

Chi-square statistic: 22.72
It measures the association between the two categorical variables (sex and target). A higher value suggests a stronger association between sex and target.

P-value: 1.88e-06
This p-value is also very small, indicating the likelihood of having heart disease is significantly related to the patient's sex.

It means both age and sex are important factors in predicting the presence of heart disease.

```python

numerical_cols = ['cp','trestbps','chol','thalach','oldpeak','ca','thal']
for col in numerical_cols:
    group1 = df1[df1['target'] == 0][col]
    group2 = df1[df1['target'] == 1][col]
    t_statistic, p_value = ttest_ind(group1, group2)
    print(f"\nT-test for {col}:")
    print("T-statistic:", t_statistic)
    print("P-value:", p_value)
```

```python

categorical_cols = ['fbs', 'restecg', 'exang', 'slope']
for col in categorical_cols:
    contingency_table = pd.crosstab(df1[col], df1['target'])
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    print(f"\nChi-square test for {col} and target:")
    print("Chi-square statistic:", chi2)
    print("P-value:", p)
```

T-Tests: Significant differences were found in the means for cp, trestbps, thalach, oldpeak, ca, and thal between the two groups. There was no significant difference in the mean for chol.

Chi-Square Tests: Significant associations were found between restecg, exang, and slope with the target. There was no significant association between fbs and the target.
