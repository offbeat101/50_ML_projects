```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv('movies.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
needed_cols = ['genres', 'keywords', 'tagline', 'cast', 'director']
d = df[needed_cols]
d.isnull().sum()
```

```python
for col in needed_cols:
  d[col].fillna('',inplace=True)
```

```python
d.isnull().sum()
```

```python
combined_cols = d['genres'] + " " + d['keywords'] + " " + d['tagline'] + " " + d['cast'] + " " + d['director']
combined_cols
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vec=TfidfVectorizer()
vec_matrix=vec.fit_transform(combined_cols)
vec_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vec_matrix)
similarity
```

```python
movies = df['title'].tolist()
movies
```

```python
import difflib
user_input = input("Enter the movie: ")
movie_names = difflib.get_close_matches(user_input, movies)
print(movie_names)
```

```python
close_match = movie_names[0]
print(close_match)
```

```python
index = df[df.title == close_match]['index'].values[0]
print(index)
```

```python
similarity_score = list(enumerate(similarity[index]))
print(similarity_score)
```

```python
sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)
print(sorted_similar_movies)
```

```python
print('Movies suggested for you : \n')

i = 1

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = df[df.index==index]['title'].values[0]
  if (i<30):
    print(i, '.',title_from_index)
    i+=1
```
