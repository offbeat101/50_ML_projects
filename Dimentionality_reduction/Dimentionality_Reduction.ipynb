```python


from sklearn import datasets
iris = datasets.load_iris()
```

```python
import pandas as pd
import plotly.express as px
```

```python
df=px.data.iris()
df.head
```

```python
df
```

```python
features=['sepal_length','sepal_width','petal_length','petal_width']
fig=px.scatter_matrix(df,dimensions=features,color='species')
fig.show()
```

```python
from sklearn.manifold import TSNE
features=df.loc[:,:'petal_width']
model=TSNE(n_components=2,random_state=7)
tsne_features=model.fit_transform(features)
fig=px.scatter(tsne_features,x=0,y=1,color=df['species'])
fig.show()
```

```python

features=['sepal_length','sepal_width','petal_length','petal_width']
fig=px.scatter_matrix(df,dimensions=features,color='species')
fig.show()
features=df.loc[:,:'petal_width']
model=TSNE(n_components=3,random_state=7)
tsne_features=model.fit_transform(features)

fig = px.scatter_3d(tsne_features, x=0, y=1, z=2, color=df['species'])
fig.show()
```
