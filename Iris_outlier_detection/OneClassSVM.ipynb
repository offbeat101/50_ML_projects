```python
from sklearn import datasets

iris = datasets.load_iris()
```

```python
import seaborn as sns
import pandas as pd
```

```python
df=sns.load_dataset('iris')
df.head()
```

```python
sns.scatterplot(df,x='sepal_length',y='sepal_width',hue='species')
```

```python
sns.scatterplot(df,x='petal_length',y='petal_width',hue='species')
```

```python
from sklearn.svm import OneClassSVM
model=OneClassSVM(nu=0.001)
x=df[['sepal_width','petal_width']]

model.fit(x)
model.predict(x)
```

```python
sns.scatterplot(df,x='sepal_width',y='petal_width',hue=model.predict(x))
```

```python

from sklearn.ensemble import IsolationForest
model = IsolationForest(contamination=0.01)
model.fit(x)
sns.scatterplot(df,x='sepal_width',y='petal_width',hue=model.predict(x))
```

```python


from sklearn.neighbors import LocalOutlierFactor

model = LocalOutlierFactor(n_neighbors=20, contamination=0.01)
y_pred = model.fit_predict(x)
sns.scatterplot(df,x='sepal_width',y='petal_width',hue=y_pred)
```
